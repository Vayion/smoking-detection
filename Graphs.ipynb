```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


df = pd.read_csv("./dataset/dataset.csv")
```

```python
smokers = df[df["smoking"]==1]
clean = df[df["smoking"]==0]
```

```python
label_x = "age"
label_y = "height(cm)"
```

```python
label_x = "age"
label_y = "weight(kg)"
```

```python
label_x = "age"
label_y = "fasting blood sugar"
```

```python
label_x = "weight(kg)"
label_y = "fasting blood sugar"
```

```python
avg_smokers = smokers.groupby(label_x)[label_y].mean()
avg_clean = clean.groupby(label_x)[label_y].mean()

fig, ax = plt.subplots()

ax.plot(avg_smokers.index, avg_smokers.values, linewidth=2.0, label = "smokers")
ax.plot(avg_clean.index, avg_clean.values, linewidth=2.0, label = "non smokers")
ax.set_xlabel(label_x)
ax.set_ylabel(label_y)
ax.legend()
```

```python
tg_smoker = smokers["triglyceride"]
tg_nonsmoker = clean["triglyceride"]

log_tg_smoker = np.log1p(tg_smoker)
log_tg_nonsmoker = np.log1p(tg_nonsmoker)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(tg_nonsmoker, bins=50, alpha=0.6, label="Non-smokers")
plt.hist(tg_smoker, bins=50, alpha=0.6, label="Smokers")
plt.xlabel("Triglyceride")
plt.ylabel("Count")
plt.title("Raw Triglycerides")
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(log_tg_nonsmoker, bins=50, alpha=0.6, label="Non-smokers")
plt.hist(log_tg_smoker, bins=50, alpha=0.6, label="Smokers")
plt.xlabel("log(Triglyceride + 1)")
plt.ylabel("Count")
plt.title("Log-Transformed Triglycerides")
plt.legend()

plt.tight_layout()
plt.show()
```
