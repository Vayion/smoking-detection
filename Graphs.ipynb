```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


df = pd.read_csv("./dataset/dataset.csv")
```

```python
smokers = df[df["smoking"]==1]
clean = df[df["smoking"]==0]
```

```python
label_x = "age"
label_y = "height(cm)"
```

```python
label_x = "weight(kg)"
label_y = "fasting blood sugar"
```

```python
avg_smokers = smokers.groupby(label_x)[label_y].mean()
avg_clean = clean.groupby(label_x)[label_y].mean()

fig, ax = plt.subplots()

ax.plot(avg_smokers.index, avg_smokers.values, linewidth=2.0, label = "smokers")
ax.plot(avg_clean.index, avg_clean.values, linewidth=2.0, label = "non smokers")
ax.set_xlabel(label_x)
ax.set_ylabel(label_y)
ax.legend()
```
