```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("./dataset/dataset.csv")
z = df.pop("gender")
y = df.pop("smoking").astype(int)
X = df
```

```python
import pandas as pd
import statsmodels.api as sm
from sklearn.preprocessing import RobustScaler


#Robust Scaler to ignore all extreme outliers
scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)
X = pd.DataFrame(
    X_scaled,
    columns = X.columns,
    index = X.index
)

X = sm.add_constant(X)

model = sm.Logit(y, X)
result = model.fit()
print(result.summary())
```
