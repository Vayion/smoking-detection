```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("./dataset/dataset.csv")

df_men = df[df["gender"]==1]

df_women = df[df["gender"]==0]
df_women = df_women.drop(columns = "gender")
df_men = df_men.drop(columns = "gender")
```

```python
import statsmodels.api as sm
from sklearn.preprocessing import RobustScaler


#Robust Scaler to ignore all extreme outliers
def significance_of(X, y):
    scaler = RobustScaler()
    X_scaled = scaler.fit_transform(X)
    X = pd.DataFrame(
        X_scaled,
        columns = X.columns,
        index = X.index
    )

    X = sm.add_constant(X)

    model = sm.Logit(y, X)
    result = model.fit()
    print(result.summary())
```

```python
y = df_women["smoking"].copy()
X = df_women.drop(columns=["smoking"]).copy()

significance_of(X, y)
```

```python
y = df_men["smoking"].copy()
X = df_men.drop(columns=["smoking"]).copy()

significance_of(X, y)
```
