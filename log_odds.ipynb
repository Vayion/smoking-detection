```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("./dataset/dataset.csv")
y = df.pop("smoking").astype(int)
X = df
```

```python
def log_odds_plot(x, bins=10):
    df = pd.DataFrame({x: X[x], "smoking": y})
    df["bin"] = pd.qcut(df[x], bins, duplicates="drop")

    grouped = df.groupby("bin", observed=True)["smoking"].mean()
    log_odds = np.log(grouped / (1 - grouped))

    plt.plot(range(len(log_odds)), log_odds, marker="o")
    plt.title("Log-Odds vs. Quantile: "+ x)
    plt.xlabel("Quantil-Bin")
    plt.ylabel("Log-Odds")
    plt.show()
```

```python
for column in X:
    log_odds_plot(column)
```
