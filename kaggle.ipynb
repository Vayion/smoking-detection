```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("kukuroo3/body-signal-of-smoking")

print("Path to dataset files:", path)
```

```python
import pandas as pd

raw_data = pd.read_csv(path+"/smoking.csv")
numeric_data = pd.read_csv(path+"/smoking.csv")
numeric_data = numeric_data.drop("gender", axis = 1)
numeric_data = numeric_data.drop("oral", axis = 1)
numeric_data = numeric_data.drop("tartar", axis = 1)
```

```python

raw_data.dtypes
```

```python
smokers = numeric_data[numeric_data["smoking"] == 1]
smokers.mean()
```

```python
non_smokers = numeric_data[numeric_data["smoking"] == 0]
non_smokers.mean()
```

```python
female = raw_data[raw_data["gender"] == "F"]
female = female.drop("gender", axis = 1)
female = female.drop("oral", axis = 1)
female = female.drop("tartar", axis = 1)
female.mean()
```

```python
male = raw_data[raw_data["gender"] == "M"]
male = male.drop("gender", axis = 1)
male = male.drop("oral", axis = 1)
male = male.drop("tartar", axis = 1)
male.mean()
```
